```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set()
```

```python
trn = pd.read_csv("data/train.csv")
trn.head()
```

```python
sexyboy = trn.corr(numeric_only=True)["SalePrice"].drop("SalePrice")
plt.figure(figsize=(12,2))
plt.bar(sexyboy.index,sexyboy)
plt.ylabel("Correlation!")
plt.xticks(rotation = 90)
plt.show()
```
